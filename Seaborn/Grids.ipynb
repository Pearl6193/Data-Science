```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.info()
```

```python
iris.head()
```

```python
iris["species"].unique()
```

```python
g=sns.PairGrid(iris)
```

```python
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue="species",palette="plasma")
```

```python
sns.pairplot(iris,hue="species",diag_kind="hist")
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
g=sns.FacetGrid(tips,row="sex",col="smoker")
```

```python
g=sns.FacetGrid(tips,row="sex",col="smoker")
g.map(plt.hist,"tip")
```

```python
g=sns.FacetGrid(tips,row="sex",col="smoker",hue="time")
g.map(plt.scatter,"tip","total_bill").add_legend()
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill",y="tip",data=tips)
g = g.plot(sns.regplot,sns.kdeplot)
```

```python
g = sns.JointGrid(x="total_bill",y="tip",data=tips)
g = g.plot(sns.regplot,sns.histplot)
```
