```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips)
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="plasma")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="plasma",markers=["o","v"],scatter_kws={'s':100})
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="magma",col="sex")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="magma",col="sex",row="time")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="magma",col="day")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",palette="magma",col="day",aspect=0.5,size=8)
```
